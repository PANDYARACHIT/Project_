```python
def parquet_to_csv(parquet_file, csv_file):
    # Read Parquet file into a DataFrame
    df = pd.read_parquet(parquet_file)

    # Write DataFrame to CSV
    df.to_csv(csv_file, index=False)
```

```python
# File names  
fhvhv_tripdata_file = 'fhvhv_tripdata_2022-12.parquet'
fhv_tripdata_file = 'fhv_tripdata_2022-12 (1).parquet'
yellow_tripdata_file = 'yellow_tripdata_2022-12 (1).parquet'
green_tripdata_file = 'green_tripdata_2022-12 (1).parquet'

# Output CSV file names
fhvhv_tripdata_csv = 'fhvhv_tripdata_2022-12.csv'
fhv_tripdata_csv = 'fhv_tripdata_2022-12.csv'
yellow_tripdata_csv = 'yellow_tripdata_2022-12.csv'
green_tripdata_csv = 'green_tripdata_2022-12.csv'

# Convert Parquet to CSV
parquet_to_csv(fhvhv_tripdata_file, fhvhv_tripdata_csv)
parquet_to_csv(fhv_tripdata_file, fhv_tripdata_csv)
parquet_to_csv(yellow_tripdata_file, yellow_tripdata_csv)
parquet_to_csv(green_tripdata_file, green_tripdata_csv)
```

```python
import sys

yellow_df = pd.read_csv('yellow_taxi_data.csv')

# Setting the maximum memory size in bytes so that we can load it in Looker
max_memory_size = 10779361

# Initialize a variable to store cumulative memory usage
cumulative_memory = 0

# Iterate through rows and check memory usage for yellow_df
for idx, row in yellow_df.iterrows():
    # Calculate memory usage for the current row
    row_memory = sum(sys.getsizeof(value) if not pd.api.types.is_integer_dtype(type(value)) else 0 for value in row)
    
    # Check if adding the current row plus CSV overhead exceeds the maximum memory size
    if cumulative_memory + row_memory + sys.getsizeof('\n') > max_memory_size:
        break
    
    # Add the current row's memory usage to the cumulative memory
    cumulative_memory += row_memory

# Create a new DataFrame with the selected rows for yellow_df
selected_rows_yellow = yellow_df.head(idx+1)

# Save the selected DataFrame to a new CSV file for yellow_df
selected_rows_yellow.to_csv('selected_yellow_taxi_data.csv', index=False)
```
